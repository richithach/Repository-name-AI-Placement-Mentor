```python
import os
import google.generativeai as genai

# Configure Gemini API
# Set your API key here temporarily while testing
os.environ["GOOGLE_API_KEY"] = "YOUR_ACTUAL_NEW_KEY"

genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))

# Create model
model = genai.GenerativeModel(
    "gemini-2.5-flash",
    system_instruction="""
    You are AI Placement Mentor.

    Your purpose is to help engineering students with:
    - Resume reviews
    - Interview preparation
    - DSA roadmaps
    - Python learning
    - SQL learning
    - AI and Data Analytics career guidance
    - Project suggestions
    - Placement preparation

    Give practical, actionable and beginner-friendly advice.
    """
)

# Start chat
chat = model.start_chat(history=[])

print("=" * 50)
print("       AI PLACEMENT MENTOR")
print("=" * 50)

while True:

    print("\n===== MENU =====")
    print("1. General Chat")
    print("2. Interview Questions")
    print("3. Resume Review")
    print("4. Skill Roadmap")
    print("5. Project Suggestions")
    print("6. Exit")

    choice = input("\nChoose option: ")

    try:

        # General Chat
        if choice == "1":
            question = input("\nAsk anything: ")

            response = chat.send_message(question)

            print("\nAssistant:")
            print(response.text)

        # Interview Questions
        elif choice == "2":
            role = input("\nEnter role: ")

            prompt = f"""
            Generate 10 interview questions for a fresher applying for {role}.

            Include:
            - Beginner questions
            - Intermediate questions
            - Short answers
            """

            response = chat.send_message(prompt)

            print("\nAssistant:")
            print(response.text)

        # Resume Review
        elif choice == "3":
            resume = input("\nPaste your resume text:\n")

            prompt = f"""
            Review this resume.

            Give:
            1. Strengths
            2. Weaknesses
            3. ATS Score Estimate
            4. Suggested Improvements
            5. Missing Skills

            Resume:
            {resume}
            """

            response = chat.send_message(prompt)

            print("\nAssistant:")
            print(response.text)

        # Skill Roadmap
        elif choice == "4":
            skill = input("\nEnter skill or role: ")

            prompt = f"""
            Create a detailed 30-day learning roadmap for {skill}.

            Include:
            - Daily goals
            - Resources
            - Practice suggestions
            - Mini projects
            """

            response = chat.send_message(prompt)

            print("\nAssistant:")
            print(response.text)

        # Project Suggestions
        elif choice == "5":
            role = input("\nEnter target role: ")

            prompt = f"""
            Suggest 5 resume-worthy projects for {role}.

            For each project provide:
            1. Project Name
            2. Technologies Used
            3. Difficulty Level
            4. Skills Learned
            5. Why it is valuable on a resume
            """

            response = chat.send_message(prompt)

            print("\nAssistant:")
            print(response.text)

        # Exit
        elif choice == "6":
            print("\nThank you for using AI Placement Mentor!")
            break

        else:
            print("\nInvalid option. Please try again.")

    except Exception as e:
        print("\nError:", e)
```
